# The Massachusetts Institute of Technology (MIT)

**Nina Speller / NinaSpeller4**

The Massachusetts Institute of Technology (MIT) is a private university in Cambridge, Massachusetts. Established in 1861, MIT has since then been at the center of the development of modern technology and science. Since it was founded, MIT has been ranked among the top academic institutions in the world.

The historic (original) MIT seal (logo) design was recommended by a committee that included founder William Barton Rogers and approved by MIT’s governing board in 1864. No single individual artist is officially recorded as its creator.

The two symbols in the original MIT logo represent the Scholar and the Artisan, illustrating MIT’s philosophy of uniting theoretical knowledge with practical application, as expressed in its motto Mens et Manus (Mind and Hand).
